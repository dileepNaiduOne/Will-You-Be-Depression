```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
import numpy as np
import random

# Define the number of rows
num_rows = 413768

# Define the column names
columns = [
    "Name", "Age", "Marital Status", "Education Level",
    "Number of Children", "Smoking Status", "Physical Activity Level",
    "Employment Status", "Income", "Alcohol Consumption",
    "Dietary Habits", "Sleep Patterns", "History of Mental Illness",
    "History of Substance Abuse", "Family History of Depression",
    "Chronic Medical Conditions" 
]

# Create an empty DataFrame
df = pd.DataFrame(columns=columns)

# 1. Name Generation:
names = [
    "Aakash", "Priya", "Rahul", "Anjali", "Amit", "Smita", 
    "Rajeev", "Neha", "Vivek", "Deepika", "Siddharth", "Shruti",
    "Karan", "Avani", "Rohit", "Rhea", "Ashish", "Aditi", 
    "Sachin", "Simran", "Vikram", "Pooja", "Praveen", "Shilpa",
    "Mohan", "Reena", "Nilesh", "Sarika", "Vinod", "Manisha",
    "Dinesh", "Rani", "Harish", "Meena", "Sunil", "Jyoti",
    "Akash", "Rina", "Raj", "Ankita", "Ashwin", "Anusha",
    "Vishal", "Renu", "Nitin", "Sonam", "Manoj", "Swati",
    "Mukesh", "Nisha", "Pawan", "Sanjana", "Ajay", "Divya",
    "Ajay", "Ankita", "Arun", "Bharti", "Deepak", "Geeta",
    "Harish", "Kiran", "Madhu", "Mohan", "Nisha", "Om",
    "Pawan", "Ramesh", "Sohan", "Usha", "Vikram", "Yash",
    "Ajay", "Anju", "Ashish", "Bhavna", "Dinesh", "Gaurav",
    "Hari", "Kalyani", "Madhav", "Nayan", "Omkar", "Pradeep",
    "Ram", "Sanjay", "Uday", "Varsha", "Vijay", "Zahir",
]

df["Name"] = random.choices(names, k=num_rows)

# 2. Age Generation:
df["Age"] = np.random.randint(18, 65, size=num_rows)

# 3. Marital Status Generation:
marital_statuses = ["Married", "Single", "Widowed", "Divorced"]
df["Marital Status"] = np.random.choice(marital_statuses, size=num_rows)

# 4. Education Level Generation:
education_levels = ["High School", "Bachelor's Degree", "Master's Degree", "Associate Degree", "PhD"]
df["Education Level"] = np.random.choice(education_levels, size=num_rows)

# 5. Number of Children Generation:
df["Number of Children"] = np.random.randint(0, 5, size=num_rows)

# 6. Smoking Status Generation:
smoking_statuses = ["Non-smoker", "Former", "Current"]
df["Smoking Status"] = np.random.choice(smoking_statuses, size=num_rows)

# 7. Physical Activity Level Generation:
activity_levels = ["Active", "Sedentary", "Moderate"]
df["Physical Activity Level"] = np.random.choice(activity_levels, size=num_rows)

# 8. Employment Status Generation:
employment_statuses = ["Employed", "Unemployed"]
df["Employment Status"] = np.random.choice(employment_statuses, size=num_rows)

# 9. Income Generation:
df["Income"] = np.random.randint(10000, 100000, size=num_rows)

# 10. Alcohol Consumption Generation:
alcohol_consumption = ["Moderate", "High", "Low"]
df["Alcohol Consumption"] = np.random.choice(alcohol_consumption, size=num_rows)

# 11. Dietary Habits Generation:
dietary_habits = ["Healthy", "Moderate", "Unhealthy"]
df["Dietary Habits"] = np.random.choice(dietary_habits, size=num_rows)

# 12. Sleep Patterns Generation:
sleep_patterns = ["Good", "Fair", "Poor"]
df["Sleep Patterns"] = np.random.choice(sleep_patterns, size=num_rows)

# 13. History of Mental Illness Generation:
df["History of Mental Illness"] = np.random.choice(["Yes", "No"], size=num_rows, p=[0.2, 0.8])  # Adjust probabilities as needed

# 14. History of Substance Abuse Generation:
df["History of Substance Abuse"] = np.random.choice(["Yes", "No"], size=num_rows, p=[0.1, 0.9])  # Adjust probabilities as needed

# 15. Family History of Depression Generation:
df["Family History of Depression"] = np.random.choice(["Yes", "No"], size=num_rows, p=[0.3, 0.7])  # Adjust probabilities as needed

# 16. Chronic Medical Conditions Generation (Depression):
#    - Implement correlations using conditional probabilities:
df["Chronic Medical Conditions"] = "No"  # Default to "No" for depression
for i in range(num_rows):
    if df.loc[i, "Physical Activity Level"] == "Sedentary":
        df.loc[i, "Chronic Medical Conditions"] = np.random.choice(["Yes", "No"], p=[0.6, 0.4])
    if df.loc[i, "Sleep Patterns"] == "Poor":
        df.loc[i, "Chronic Medical Conditions"] = np.random.choice(["Yes", "No"], p=[0.9, 0.1])
    if df.loc[i, "Smoking Status"] == "Current":
        df.loc[i, "Chronic Medical Conditions"] = np.random.choice(["Yes", "No"], p=[0.7, 0.3])

# 17. Save the Dataset:
df.to_csv("depression_dataset.csv", index=False)

print("Dataset generated and saved as 'depression_dataset.csv'")
```

```python
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(20, 7))

for i in range(13):
        
    feature = ['Marital Status', 'Education Level', 'Number of Children',
       'Smoking Status', 'Physical Activity Level', 'Employment Status',
       'Alcohol Consumption', 'Dietary Habits', 'Sleep Patterns',
       'History of Mental Illness', 'History of Substance Abuse',
       'Family History of Depression', 'Depression']
    
    category = df[feature[i]].value_counts().index
    percentage = [f"{i:.2f}%" for i in (df[feature[i]].value_counts().values/df.shape[0])*100]
    
    f_table = pd.DataFrame({"Category":category, "Percentage":percentage})

    axes[i//4, i%4].axis('tight')
    axes[i//4, i%4].axis('off')
    axes[i//4, i%4].table(cellText=f_table.values, colLabels=f_table.columns, loc="center", cellLoc="center", fontsize=15, colColours=["yellow"]*13)
    axes[i//4, i%4].set_title(feature[i])



axes[3, 1].axis('off')
axes[3, 2].axis('off')
axes[3, 3].axis('off')

plt.tight_layout()
plt.show()
```
